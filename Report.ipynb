## Creating the Baseline

Before we start desiging and modelling our own policies with regards to buying and selling players, we need to investigate what the best outcome could be.
Hence, the optimal solution to the policy: "maximizing rate of return" is calculated, which will serve as upper bound to any potential investment strategy.

The problem was solved by iteratively solving the tail-problem, starting from the final time $T_\text{Final}$ and moving back in time towards the initial time $T_0$. The figure below shows three plots: Valuation of Squad, Cash Reserves, and Total Club Wealth, which are all obtained from the optimal solution.

The first figure shows the combined value of all the players in the squad at the start and at the end of each year. The plot shows that every year a good profit is made be selling the roster, and investing in new players.

The second plot shows how the cash reserves of the club evolve over time. The cash reverses build up each year, because the investment in new players each year is consistently lower than the price for which the old players are sold.

The third plot shows the total club wealth, which is made up of the value of the players currently employed and the built-up cash reserves. The plots shows that as time progresses, the wealth of the club increases, where the initial wealth of the club was equal to the initial budget $b_0$.


![picture](
  https://drive.google.com/uc?export=view&id=1h8miab3k_py1GA61H48hKtDKGaros5j4)
Fig. 1: Optimal Policy: Maximize Rate of Return



In order to be able to quantify how well the optimal solution performs, the retrospective rate of return is calculated for the obtained club wealth. This is done using the equation below where $b_0$ denotes the initial budget, $b_T$ the final club wealth, $T$ the time horizon of 10 years, and $r$ the rate of return.

\begin{equation}
\begin{aligned}
b_T &= b_0 r ^T\\
r &= (\frac{b_T}{b_0})^{\frac{1}{T}} \approx 1.52
\end{aligned}
\end{equation}

The equation above shows that the optimal solution is able to acquire a whopping compound interest of 52% annualy.
